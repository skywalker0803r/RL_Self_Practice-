```
import imageio
import numpy as np
from stable_baselines import PPO2
from stable_baselines.common.cmd_util import make_atari_env
from stable_baselines.common.vec_env import VecFrameStack

# create env
env_name = 'BoxingNoFrameskip-v0'
env = make_atari_env(env_name,num_env=4,seed=42)
env = VecFrameStack(env,n_stack=4)

# create model
model = PPO2("CnnPolicy",env,verbose=True).learn(100000)
```

```
model.learn(10000)
```

```
model.save(save_path='/content/drive/MyDrive/RL MODEL/{}_{}'.format(model.__class__.__name__,env_name))
print('save ok')
```

```
del model
```

```
env = make_atari_env(env_name,num_env=4,seed=42)
env = VecFrameStack(env,n_stack=4)
model = PPO2.load('/content/drive/MyDrive/RL MODEL/PPO2_BoxingNoFrameskip-v0.zip',env=env)
```

```
model.learn(10000)
```

# watch

```
from tqdm import tqdm_notebook as tqdm

# num_env = 1
env = make_atari_env(env_name,num_env=1,seed=42)
env = VecFrameStack(env,n_stack=4)

images = []
obs = env.reset()
img = env.render(mode='rgb_array')
for i in tqdm(range(1000)):
    images.append(img)
    action, _ = model.predict(obs)
    obs, _, _ ,_ = env.step(action)
    img = env.render(mode='rgb_array')

imageio.mimsave('{}.gif'.format(env_name), [np.array(img) for i, img in enumerate(images) if i%2 == 0], fps=29)
```
